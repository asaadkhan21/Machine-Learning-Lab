```python
#Line Plot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.plot(x,y)
plt.show()
```

```python
#Bar Plot
dat=data = {'C':20, 'C++':15, 'Java':30,'Python':35}
course=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(10,5))
plt.bar(course,values,color='green', width=0.5)
plt.title("Students Enrolled in Different Courses", color='red')
plt.xlabel("Courese Offered")
plt.ylabel("Number of Students Enrolled")
plt.show()
```

```python
#Box Plot
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
# Creating plot
plt.boxplot(data)
plt.show()
```

```python
#Scatter Plot
x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]
y =[99, 86, 87, 88, 100, 86,
    103, 87, 94, 78, 77, 85, 86]
plt.scatter(x, y, c ="blue")
# To show the plot
plt.show()
```

```python
#Pie Chart
import seaborn 
data = [44, 45, 40, 41, 39]
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5']
# define Seaborn color palette to use
palette_color = seaborn.color_palette('bright') 
# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,autopct='%.0f%%' ) 
# displaying chart
plt.show()
```
