```python
#Line Plot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.plot(x,y)
plt.show()
```

```python
#Bar Plot
dat=data = {'C':20, 'C++':15, 'Java':30,'Python':35}
course=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(10,5))
plt.bar(course,values,color='green')
plt.title("Students Enrolled in Different Courses", color='red')
plt.xlabel("Courese Offered")
plt.ylabel("Number of Students Enrolled")
plt.show()
```
